## 1. If statements

`if` lets Python choose what to do.

```python
age = 15

if age >= 13:
    print("You are a teenager")
```

## 2. Else

`else` runs if the condition is not true.

```python
age = 10

if age >= 13:
    print("You are a teenager")
else:
    print("You are not a teenager yet")
```

## 5. Elif

`elif` means “else if”.

```python
score = 75

if score >= 90:
    print("Excellent")
elif score >= 60:
    print("Good")
else:
    print("Keep practising")
```
