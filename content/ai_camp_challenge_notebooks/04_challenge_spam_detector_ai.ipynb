# 📧 Challenge 1: Build a Spam Detector AI

You work for an email company. Your job is to build a tiny AI that decides whether an email is **Spam** or **Not Spam**.

This is not a real modern AI yet — you are manually designing the “brain” by choosing the weights and thresholds.

## How the AI works

Each email is described by four features:

1. Contains the word **FREE**
2. Contains the phrase **CLICK HERE**
3. Has lots of **CAPITAL LETTERS**
4. Is from a **known sender**

Each feature is either:

- `1` = yes
- `0` = no

The AI has two output neurons:

- Output 1 = Spam
- Output 2 = Not Spam


````python
import numpy as np

feature_names = [
    "Contains FREE",
    "Contains CLICK HERE",
    "Lots of CAPITAL LETTERS",
    "Known sender"
]

output_names = ["Spam", "Not Spam"]

def display_email(features):
    print("Email features:")
    for name, value in zip(feature_names, features):
        symbol = "✅" if value == 1 else "❌"
        print(f"{symbol} {name}")

def run_ai(features, weights, thresholds, show=True):
    features = np.array(features)
    output_values = features @ weights

    active_outputs = []
    for i, value in enumerate(output_values):
        if value >= thresholds[i]:
            active_outputs.append(output_names[i])

    if show:
        display_email(features)
        print("\nOutput values:")
        for name, value, threshold in zip(output_names, output_values, thresholds):
            print(f"{name}: {value:.2f}  (threshold: {threshold})")
        print("\nPrediction:")
        if len(active_outputs) == 0:
            print("No output neuron fired.")
        elif len(active_outputs) == 1:
            print(active_outputs[0])
        else:
            print("Both fired:", active_outputs)

    return active_outputs, output_values
````

## Part 1 — Try one email

Change the `email` list and see what happens.

The order is:

```text
[FREE, CLICK HERE, CAPITAL LETTERS, Known sender]
```


````python
email = [1, 1, 1, 0]

weights = np.array([
    [3.0, 0.0],    # FREE              -> [Spam, Not Spam]
    [3.0, 0.0],    # CLICK HERE        -> [Spam, Not Spam]
    [2.0, 0.0],    # CAPITAL LETTERS   -> [Spam, Not Spam]
    [-3.0, 3.0],   # Known sender      -> [Spam, Not Spam]
])

thresholds = np.array([
    4.0,   # Spam threshold
    2.0    # Not Spam threshold
])

run_ai(email, weights, thresholds)
````

## Part 2 — Test examples

Run the test set below. Your current AI will probably make some mistakes.

Your goal is to edit the **weights** and **thresholds** to get as many correct as possible.


````python
test_emails = [
    # features, correct answer
    ([1, 1, 1, 0], "Spam"),
    ([1, 0, 1, 0], "Spam"),
    ([0, 1, 1, 0], "Spam"),
    ([0, 0, 0, 1], "Not Spam"),
    ([1, 0, 0, 1], "Not Spam"),
    ([0, 1, 0, 1], "Not Spam"),
    ([0, 0, 1, 1], "Not Spam"),
    ([1, 1, 0, 1], "Spam"),       # tricky
    ([0, 0, 0, 0], "Not Spam"),   # tricky: unknown sender but no suspicious words
]

def predict_label(features, weights, thresholds):
    active_outputs, output_values = run_ai(features, weights, thresholds, show=False)

    if len(active_outputs) == 1:
        return active_outputs[0]
    elif len(active_outputs) == 0:
        return "No decision"
    else:
        return "Confused"

def test_ai(weights, thresholds):
    correct = 0

    for features, answer in test_emails:
        prediction = predict_label(features, weights, thresholds)
        mark = "✅" if prediction == answer else "❌"
        if prediction == answer:
            correct += 1
        print(f"{mark} Features {features} | predicted: {prediction:12s} | correct: {answer}")

    print(f"\nScore: {correct}/{len(test_emails)}")

test_ai(weights, thresholds)
````
